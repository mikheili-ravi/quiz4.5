```python
import sqlite3
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
#
connection = sqlite3.connect("pokemon_data.sqlite3")
cursor = connection.cursor()

#  fire  
query = "SELECT * FROM pokemon_data WHERE field6 = 'fire';"
cursor.execute(query)

col = cursor.fetchall()

for row in col:
    print(row)

# #   
# name = input("  : ")
# type = input(" type: ")
# weight = int(input(" weight: "))

# cursor.execute("INSERT INTO pokemon_data (field2, field6, field5) VALUES (?, ?, ?);", (name, type, weight))
# connection.commit()

# #   
# name = input("  ,    : ")
# new_weight = int(input("  : "))

# cursor.execute("UPDATE pokemon_data SET field5 = ? WHERE field2 = ?;", (new_weight, name))
# connection.commit()

# print(f" : {new_weight}")

# #    
# name_to_delete = input("   : ")

# cursor.execute("DELETE FROM pokemon_data WHERE field2 = ?;", (name_to_delete,))
# connection.commit()

# print(f" : {name_to_delete}")



#   a
cursor.execute("SELECT field6 FROM pokemon_data")
rows = cursor.fetchall()
all_types = []
for row in rows:
    cell = row[0]
    if cell:
        for t in cell.split(','):
            kemon = t.strip().lower()
            if "type" not in kemon:
                all_types.append(kemon)

type_counts = Counter(all_types)

labels = list(type_counts.keys())
values = list(type_counts.values())

plt.figure(figsize=(10,5))
plt.bar(labels, values, color='red')
plt.title("  ")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



plt.style.use('_mpl-gallery-nogrid')

conn = sqlite3.connect('pokemon_data.sqlite3')
cursor = conn.cursor()

cursor.execute("SELECT field6 FROM pokemon_data")
rows = cursor.fetchall()[1:] 

allowed_types = {'water', 'fire', 'ground', 'normal'}
type_counts = {}

for row in rows:
    primary_type = row[0].split(',')[0].strip().lower()
    if primary_type in allowed_types:
        type_counts[primary_type] = type_counts.get(primary_type, 0) + 1

labels = list(type_counts.keys())
sizes = list(type_counts.values())
colors = plt.get_cmap('tab10')(np.linspace(0.1, 0.9, len(labels)))

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)
ax.set(xlim=(0, 8), ylim=(0, 8))
plt.show()

plt.style.use('_mpl-gallery')


#  
cursor.execute("SELECT field5, field6 FROM pokemon_data")
rows = cursor.fetchall()[1:]      

types_to_show = ['fire', 'ice', 'normal']
type_weights  = {t: [] for t in types_to_show}

for weight_str, type_str in rows:
    primary = type_str.split(',')[0].strip().lower()
    if primary in type_weights:
        type_weights[primary].append(float(weight_str))

min_len = min(len(v) for v in type_weights.values())
weights = [type_weights[t][:min_len] for t in types_to_show]

x = np.arange(min_len)
y = np.vstack(weights)

colors = ['#ff5733', 
          '#33c1ff', 
          '#a3ff33']

fig, ax = plt.subplots()
ax.stackplot(x, y, colors=colors)

ax.set_axis_off()
plt.title('Fire, Ice & Normal Pokmon  ')

plt.show()


```
